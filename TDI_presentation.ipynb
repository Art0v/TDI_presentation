# 4. Modeling first model

```python
#Regression analysis to predict the price
#X = revenue_sf.loc[:,['neighbourhood_cleansed','room_type', 'accommodates']]
X = revenue_sf.loc[:,['neighbourhood_cleansed','latitude','longitude', 'room_type', 'accommodates']]

#X = revenue_sf.loc[:,['latitude','longitude','room_type', 'accommodates']]
y = revenue_sf['price']
```

```python
#Label encoding
X = pd.get_dummies(X, columns=['neighbourhood_cleansed', 'room_type'])
#X = pd.get_dummies(X, columns=['room_type'])
X.dropna(inplace=True)
```

```python
X = X.astype('float64')
```

```python
#X.loc[0]
```

```python
y.loc[0]
```

```python
#X.info()
```

```python
#splitting the dataset into test and training data
X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size = 0.2 , random_state= 0)
```

```python
print('Dimensions of the training feature matrix: {}'.format(X_train.shape))
print('Dimensions of the training target vector: {}'.format(y_train.shape))
print('Dimensions of the test feature matrix: {}'.format(X_test.shape))
print('Dimensions of the test target vector: {}'.format(y_test.shape))
```

```python
#scaler = StandardScaler()
#X_train = scaler.fit_transform(X_train)
#X_test = scaler.fit_transform(X_test)
```

```python
model_1=LinearRegression()
model_1.fit(X_train,y_train)
y_pred=model_1.predict(X_test)
#R2 score
from sklearn.metrics import r2_score, mean_squared_error
print("R2 score: ",r2_score(y_test,y_pred)*100)
print("RMSE: ",np.sqrt(mean_squared_error(y_test,y_pred)))

#Error
error_diff = pd.DataFrame({'Actual Values': np.array(y_test).flatten(), 'Predicted Values': y_pred.flatten()})
print(error_diff.head(10))

#Visualize the error
if False:
    df1 = error_diff.head(10)
    df1.plot(kind='bar',figsize=(10,7))
    plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
    plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
```

```python
#X_test
```

```python
error_diff["error"] = abs(error_diff["Actual Values"] - error_diff["Predicted Values"])
X_test["diff"] = error_diff["error"]
error_per_type = [{"room_type_Entire home/apt": X_test[X_test["room_type_Entire home/apt"]==1]["diff"].mean(),
                  "room_type_Shared room" : X_test[X_test["room_type_Shared room"]==1]["diff"].mean(),
                  "room_type_Private room": X_test[X_test["room_type_Private room"]==1]["diff"].mean()}]
#Visualize the error
df1 = error_diff.head(25)
df1.plot(kind='bar',figsize=(10,7))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

df1 =pd.DataFrame(error_per_type)
df1.plot(kind='bar',figsize=(10,7))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

X_test = X_test.drop(columns=["diff"])
```
